```python
def wc(tweets,col):
    words=''.join(tweets[col])
    colmask=np.array(Image.open("C:/Users/Abhinav/Documents/data science/term 2/data visualization/Karnataka.png"))
    STOPWORDS.update(['https','rt','co'])
    stopwords=set(STOPWORDS)
    word_mod = re.sub("[^a-zA-Z0-9_]+"," ",words)
    
    ##word_mod = ' '.join([word for word in word_mod.split()
    # if "http" not in word
    # and word!='RT'
    # and word!='co'])
    bjp_wc=WordCloud(background_color='black',mask=colmask,stopwords=stopwords,min_font_size=16).generate(word_mod)
    plt.figure(figsize=(20,10)) ###mask = themes,
    plt.imshow(bjp_wc)
    plt.tight_layout(pad=0)
    plt.axis("off")
    plt.show()
    
```

```python
wc(bjp,'text')
```
